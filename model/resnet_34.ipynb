```python
import torch
import torch.nn as nn
class ResNet_34(nn.Module):
  def __init__(self, output_dim):
    super(ResNet_34, self).__init__()

    self.n_classes = output_dim
    # 16, 1, 28, 28
    self.conv1 = nn.Sequential(
      nn.Conv2d(1,64,kernel_size = 7, stride =2), # batch , 64 ,11, 11
      nn.ReLU(),
      nn.MaxPool2d(3,stride = 2) #batch, 64 ,5 ,5
    }
    self.conv2 = nn.Sequential(
        nn.Conv2d(64,64,kernel_size = 3, stride = 1, padding = 1),
        nn.ReLU(),
        nn.Conv2d(64,64, kernel_size = 3, stride = 1, padding = 1)
    )
    self.res1 = nn.Conv2d(64,128, kernel_size = 3, stride = 1, padding = 1)
    self.conv3 = nn.Sequential(
        nn.Conv2d(128,128,kernel_size = 3, stride = 1, padding = 1),
        nn.ReLU(),
        nn.Conv2d(128,128, kernel_size = 3, stride = 1, padding = 1)
    )
    self.res2 = nn.Conv2d(128,256, kernel_size = 3, stride = 1, padding = 1)
    self.conv4 = nn.Sequential(
        nn.Conv2d(256,256,kernel_size = 3, stride = 1, padding = 1),
        nn.ReLU(),
        nn.Conv2d(256,256, kernel_size = 3, stride = 1, padding = 1)
    )
    self.res3 = nn.Conv2d(256,512, kernel_size = 3, stride = 1, padding = 1)
    self.conv5 = nn.Sequential(
        nn.Conv2d(512,512,kernel_size = 3, stride = 1, padding = 1),
        nn.ReLU(),
        nn.Conv2d(512,512, kernel_size = 3, stride = 1, padding = 1)
    )

    self.softmax = nn.Softmax(dim = -1)
    self.fc = nn.Linear(512* 5 * 5,self.n_classes)

  def forward(self,x):
    out = self.conv1(x)
    identity = out

    out = self.conv2(out)
    out = self.conv2(out)
    out = self.conv2(out)

    out = self.res1(out)
    identity = self.res1(identity)
    out += identity

    out = self.conv3(out)
    out = self.conv3(out)
    out = self.conv3(out)
    out = self.conv3(out)

    out = self.res2(out)
    identity = self.res2(identity)
    out += identity

    out = self.conv4(out)
    out = self.conv4(out)
    out = self.conv4(out)
    out = self.conv4(out)
    out = self.conv4(out)
    out = self.conv4(out)

    out = self.res3(out)
    identity = self.res3(identity)
    out += identity
    identity = out

    out = self.conv5(out)
    out = self.conv5(out)
    out = self.conv5(out)
    out += identity

    out = out.view(x,size(0),-1)
    out = self.softmax(out)
    out = self.fc(out)
    return out
```
